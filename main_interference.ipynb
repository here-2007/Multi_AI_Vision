```python
options = {
    1: {
        "name": "Object Classifier",
        "classes": ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck'],
        "model": "pernavjain/object-classifier/tensorFlow1/default",
        "img_size": (32, 32),
        "model_file": "object_classifier.h5"
    },
    2: {
        "name": "Plant Disease Classifier",
        "classes": ['Apple___Apple_scab', 'Apple___Black_rot', 'Apple___Cedar_apple_rust', 'Apple___healthy', 'Blueberry___healthy', 'Cherry___healthy', 'Cherry___Powdery_mildew', 'Corn___Cercospora_leaf_spot Gray_leaf_spot', 'Corn___Common_rust', 'Corn___healthy', 'Corn___Northern_Leaf_Blight', 'Grape___Black_rot', 'Grape___Esca_(Black_Measles)', 'Grape___healthy', 'Grape___Leaf_blight_(Isariopsis_Leaf_Spot)', 'Orange___Haunglongbing_(Citrus_greening)', 'Peach___Bacterial_spot', 'Peach___healthy', 'Pepper,_bell___Bacterial_spot', 'Pepper,_bell___healthy', 'Potato___Early_blight', 'Potato___healthy', 'Potato___Late_blight', 'Raspberry___healthy', 'Soybean___healthy', 'Squash___Powdery_mildew', 'Strawberry___healthy', 'Strawberry___Leaf_scorch', 'Tomato___Bacterial_spot', 'Tomato___Early_blight', 'Tomato___healthy', 'Tomato___Late_blight', 'Tomato___Leaf_Mold', 'Tomato___Septoria_leaf_spot', 'Tomato___Spider_mites Two-spotted_spider_mite', 'Tomato___Target_Spot', 'Tomato___Tomato_mosaic_virus', 'Tomato___Tomato_Yellow_Leaf_Curl_Virus'],
        "model": "pernavjain/plant-disease-classifier/tensorFlow1/default",
        "img_size": (256, 256),
        "model_file": "plant_disease_model.h5"
    },
    3: {
        "name": "Face Emotion Classifier",
        "classes": ['angry','disgust','fear','happy','neutral','sad','surprise'],
        "model": "pernavjain/face-emotion-classifier/tensorFlow1/default",
        "img_size": (256, 256),
        "model_file": "face_emotion_detector.h5"
    },
    4: {
        "name": "X-ray Anomaly Detection",
        "classes": ['Bacterial', 'Normal', 'Viral'],
        "model": "pernavjain/chest-x-ray-anomaly-detection/tensorFlow1/default",
        "img_size": (256, 256),
        "model_file": "augmented_chest_xray.h5"
    }
}
```

```python
print("Choose a classifier option by entering a number (1-4):")
for key, val in options.items():
    print(f"{key}) {val['name']}")
option = int(input().strip())
```

```python
if option not in options:
    print("Invalid option chosen. Exiting.")
else:
    path = kagglehub.model_download(options[option]['model'])
    model_path = f"{path}/{options[option]['model_file']}"
    print(f"Path to model files: {model_path}")
    model = load_model(model_path)

    img_path = input(f"Enter the path to the image for {options[option]['name']}: ")

    img_size = options[option]['img_size']
    img = image.load_img(img_path, target_size=img_size)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0

    print("Model input shape:", model.input_shape)
    print("Image array shape:", img_array.shape)

    predictions = model.predict(img_array)
    predicted_class = np.argmax(predictions, axis=1)
    print(f"Predicted class: {options[option]['classes'][predicted_class[0]]}")
```
